```python
class Student:

    def __init__(self, name, course = None):
        self.name = name
        self.teacher = []
        if course is None:
            self.course = []
        else:
            self.course = course
            for course in course:
                course.add_student(self)
                self.add_teacher(course.teacher)
    
    def add_course(self, course):
        if course not in self.course:
            self.course.append(course)
            course.add_student(self)
            self.add_teacher(course.teacher)
    
    def remove_course(self, course):
        if course in self.course:
            self.course.remove(course)
            course.remove_student(self)
    
    def add_teacher(self, teacher):

        if teacher not in self.teacher:
            self.teacher.append(teacher)
            teacher.add_student(self)
    
    def remove_teacher(self, teacher):

        if teacher in self.teacher:
            self.teacher.remove(teacher)
            teacher.remove_student(self)
    
    def print_course(self):
        print("student name: ", self.name)

        if not self.course: 
            print(" no course")
            return
        
        for course in self.course:
            print(" course-->", course.name)

    def print_teacher(self):
        print("student name: ", self.name)
        
        if not self.teacher: 
            print(" no teacher")
            return
        
        for teacher in self.teacher:
            print(" teacher-->", teacher.name)

class Teacher:
    def __init__(self, name, course = None):
        self.name = name
        self.student = []
        if course is None:
            self.course = []
        else:
            self.course = course
            for course in course:
                course.set_teacher(self)
                for student in course.student:
                    self.add_student(student)
    
    def add_course(self, course):

        if course not in self.course:
            self.course.append(course)
            course.teacher = self
            for student in course.student:
                    self.add_student(student)
    
    def remove_course(self, course):

        if course in self.course:
            self.course.remove(course)
            course.teacher = None

    def add_student(self, student):

        if student not in self.student:
            self.student.append(student)
            student.add_teacher(self)
            
    
    def remove_student(self, student):
        if student in self.student:
            self.course.remove(student)
            student.remove_teacher(self)
    
    def print_course(self):
        print("teacher name: ", self.name)

        if not self.course: 
            print(" no course")
            return
        
        for course in self.course:
            print(" course-->", course.name)

    def print_student(self):
        print("teacher name: ", self.name)
        
        if not self.student: 
            print(" no student")
            return
        
        for student in self.student:
            print(" student-->", student.name)

class Course:

    def __init__(self, name, teacher, student = None):
        self.name = name
        self.teacher = teacher
        if student is None:
            self.student = []
        else:
            self.student = student
        teacher.add_course(self)
    
    def add_student(self, student):
        if student not in self.student:
            self.student.append(student)
            self.teacher.add_student(student)
            student.add_course(self)
            
    
    def remove_student(self, student):
        if student in self.student:
            self.student.remove(student)
            student.remove_course(self)

    def set_teacher(self, teacher):
        self.teacher = teacher
        teacher.add_course(self)
    
    def remove_teacher(self, teacher):
        self.teacher = None
        teacher.remove_course(self)

    def print_student(self):
        print("course name:", self.name)
        if not self.student: 
            print(" no student")
            return
        
        for student in self.student:
            print(" student-->", student.name)
    
    def print_teacher(self):
        print("course name:", self.name)
        print(" teacher -->", self.teacher.name)
```
