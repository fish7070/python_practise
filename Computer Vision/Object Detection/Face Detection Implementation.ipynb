```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import cvlib as cv
from cvlib.object_detection import draw_bbox
```

```python
imPath = r'D:\Programme\py\python\Computer Vision\images\p2.jpeg'
im = cv2.imread(imPath)
plt.imshow(im[:, :, ::-1])
```

```python
bbox , conf = cv.detect_face(im)

labels = []
for i in range(len(conf)):
    labels.append('person')
```

```python
im2 = draw_bbox(im, bbox, labels, conf)
plt.imshow(im2[:, :,::-1])
```

```python
imPath = r'D:\Programme\py\python\Computer Vision\images\p.jpg'
im = cv2.imread(imPath)

bbox , labels, conf = cv.detect_common_objects(im)

im2 = draw_bbox(im, bbox, labels, conf)
plt.imshow(im2[:, :,::-1])
```
