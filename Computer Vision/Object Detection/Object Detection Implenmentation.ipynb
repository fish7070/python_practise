```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imutils
from imutils.object_detection import non_max_suppression
```

```python
imPath = r'D:\Programme\py\python\Computer Vision\images\p4.jpg'
im = cv2.imread(imPath)

hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
(r, w) = hog.detectMultiScale(im, winStride=(2, 2), padding=(4, 4), scale=1.2)
```

```python
rects = np.array([[x, y, x+w, y+h] for (x, y, w, h) in r])
rects_NMS = non_max_suppression(rects, probs=None, overlapThresh=0.6)

for (x1, y1, x2, y2) in rects_NMS:
    cv2.rectangle(im, (x1, y1), (x2, y2),(0, 0, 255), 2)
```
