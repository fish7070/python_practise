RGB to HSV conversion

```python
def fun_rgb_to_hsv(r, g, b, scaleFactor):
    r = r / 255.0;
    g = g / 255.0;
    b = b / 255.0;
    cmax = max(r, g, b);
    cmin = min(r, g, b);
    diff = cmax - cmin;
    if cmax == cmin:
        h = 0
    elif cmax == r:
        h = (60 * ((g - b) / diff) + 0) % 360;
    elif cmax == g:
        h = (60 * ((b - r) / diff) + 120) % 360;
    elif cmax == b:
        h = (60 * ((r - g) / diff) + 240) % 360;

    if h < 0:
        h = h +360;
    if cmax == 0:
        s = 0;
    else:
        s = (diff / cmax) * scaleFactor;
    v = cmax * scaleFactor;
    return h, s, v;
print(fun_rgb_to_hsv(100, 200, 50, 100));
    
    
```

```python

img = cv2.imread(r'D:\py\python\Computer Vision\Image\tulips.jpg');

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV);

lr = np.array([0, 120, 70]);
ur = np.array([15, 255, 255]);
mask1 = cv2.inRange(hsv, lr, ur);

lr = np.array([165, 120, 70]);
ur = np.array([180, 255, 255]);
mask2 = cv2.inRange(hsv, lr, ur);

mask = mask1 | mask2;

res = cv2.bitwise_and(img, img, mask=mask);

plt.figure(1);
plt.subplot(121);
plt.imshow(img[:, :, ::-1]);
plt.subplot(122);
plt.imshow(res[:, :, ::-1]);
plt.show
```
