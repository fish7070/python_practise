```python
import matplotlib as mpl
def displayImageInActualSize(I):
    dpi = mpl.rcParams['figure.dpi'];
    H, W = I.shape;
    figSize = W/float(dpi), H/float(dpi);
    fig = plt.figure(figsize = figSize);
    ax = fig.add_axes([0, 0, 1, 1]);
    ax.axis('off');
    ax.imshow(I, cmap='gray');
    plt.show;
```
